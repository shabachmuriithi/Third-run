```python
def clean_lyrics(text):
    # Remove headers like [Verse 1]
    text = re.sub(r'\[.*?\]', '', text)
    # Convert to lowercase
    text = text.lower()
    # Tokenize
    tokens = word_tokenize(text)
    # Remove punctuation and stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return tokens, text  # Return tokens for frequency, raw text for sentiment
```

```python
# Clean each song's lyrics
nasty_c_eazy_tokens, nasty_c_eazy_raw = clean_lyrics(nasty_c_eazy)
nasty_c_question_tokens, nasty_c_question_raw = clean_lyrics(nasty_c_question)
# nasty_c_sma_tokens, nasty_c_sma_raw = clean_lyrics(nasty_c_sma)
kendrick_steppers_tokens, kendrick_steppers_raw = clean_lyrics(worldwide_steppers_kendrick)
kendrick_heart_tokens, kendrick_heart_raw = clean_lyrics(the_heart_partfive_kendrick)
# kendrick_alright_tokens, kendrick_alright_raw = clean_lyrics(kendrick_alright)

# Combine lyrics per artist for aggregate analysis
nasty_c_all_tokens = nasty_c_eazy_tokens + nasty_c_question_tokens
kendrick_all_tokens = kendrick_steppers_tokens + kendrick_heart_tokens
nasty_c_all_raw = nasty_c_eazy_raw + " " + nasty_c_question_raw + " "
kendrick_all_raw = kendrick_steppers_raw + " " + kendrick_heart_raw + " "
```

```python
nasty_c_sentiment = TextBlob(nasty_c_all_raw).sentiment.polarity
kendrick_sentiment = TextBlob(kendrick_all_raw).sentiment.polarity
print("Nasty C Sentiment (Polarity):", nasty_c_sentiment)
print("Kendrick Sentiment (Polarity):", kendrick_sentiment)
```

```python
set_nasty = set(nasty_c_all_tokens)
set_kendrick = set(kendrick_all_tokens)
jaccard = len(set_nasty.intersection(set_kendrick)) / len(set_nasty.union(set_kendrick))
print("Jaccard Similarity:", jaccard)
```

```python
success_words = ['success', 'dreams', 'world', 'win', 'rich', 'money', 'blessing']
struggle_words = ['pain', 'fight', 'poor', 'hard', 'curse', 'distance']
nasty_c_success = sum(nasty_c_all_tokens.count(word) for word in success_words)
nasty_c_struggle = sum(nasty_c_all_tokens.count(word) for word in struggle_words)
kendrick_success = sum(kendrick_all_tokens.count(word) for word in success_words)
kendrick_struggle = sum(kendrick_all_tokens.count(word) for word in struggle_words)
print("Nasty C - Success:", nasty_c_success, "Struggle:", nasty_c_struggle)
print("Kendrick - Success:", kendrick_success, "Struggle:", kendrick_struggle)
```
